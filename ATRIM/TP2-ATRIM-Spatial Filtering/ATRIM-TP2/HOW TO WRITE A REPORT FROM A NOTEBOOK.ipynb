````python
f = os.path.join(IMDIR, "grass.jpg")

#Display properties
width=2
height=2

# Read and preprocess image
im = io.imread(f, as_gray=True)
im = im.astype(float)
#im = resize(im,(100,100))

# Display the original image
fig = plt.figure(figsize = (width, height))
plt.imshow(im, cmap = 'gray')
plt.title('Original')
plt.show()

# Define filter parameters
filter_sizes=[1,3,5]
sigma_values=[0.1,1,2]
k=1

# Display filters

for filtre in filter_sizes:
    for sigma in sigma_values:

        kernel = gaussianKernel(filtre, sigma)

        plt.subplot(3,6,2*k-1)
        #Display the filter
        plt.imshow(kernel, cmap = 'gray')
        plt.title('Filtre de taille '+ str(filtre) + ' et de sigma '+  str(sigma))
        plt.subplot(3,6,2*k)
        # Convolve and display the filtered image

        im_filtered_scipy = ndimage.convolve(im,kernel)
        plt.imshow(im_filtered_scipy, cmap='gray')
        plt.title('Image filtrée')
        k+=1
````
